## Valores nulos
A coluna de Cargo possui valores nulos - **`NaN`**
- `NaN` = "Not a Number" = valor nulo!
- Tome cuidado, pois valores nulos se propagam!

```python
import numpy as np

np.nan
```

```python
np.nan + 1
```

```python
np.nan * 10 + (np.nan / 2 * np.nan)
```

Para evitar problemas, o método `df.value_counts()` **ignora valores nulos** da análise.
- Esse comportamento é relativamente comum no Pandas!

## Adicionando valores nulos na análise
Podemos forçar o uso de valores nulos na análise com:

```python
df_viagens['Cargo'].value_counts(dropna=False)
```

Nada menos do que 320 mil valores nulos nos dados! Isso muda completamente nossa análise de proporção:

```python
viagens_por_cargo = (df_viagens['Cargo'].value_counts(normalize=True, dropna=False) * 100).rename('Proporção Viagens').reset_index()
```

```python
viagens_por_cargo
```

## Devemos ignorar valores nulos?
Não há uma resposta certa, cada caso é um caso.
- Se os valores nulos são raros, ou provém de dados incorretos, pode fazer sentido eliminá-los da análise.
- Se os valores nulos informarem algo (como o fato de que há um problema nos cadastros), pode ser interessante mantê-los!

### Removendo valores nulos da análise
O método `df.dropna()` remove os valores nulos de uma (ou todas) as colunas:

```python
df_viagens.dropna(subset='Cargo')
```

### Substituindo valores nulos da análise
O método `df.fillna()` preenche os valores nulos com outro valor:

```python
df_viagens.fillna("NÃO IDENTIFICADO")
```

Posso preencher os valores de uma coluna apenas:

```python
df_viagens['Cargo'] = df_viagens['Cargo'].fillna("NÃO IDENTIFICADO")
df_viagens
```
