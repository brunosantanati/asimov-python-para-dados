# Os Notebooks do Google Colab
Os notebooks são formados por **células**.

Cada célula pode conter:
- Texto simples formatado como *markdown*.
- Código Python que podemos rodar.

Podemos usar os botões da interface para:
- **Criar** novas células.
- **Modificar** células existentes.
- **Alterar a ordem** de células.
- **Deletar** células.


## "Rodando" uma célula
- Quando clicamos em uma célula, conseguimos editá-la.
- Apertando **`Shift + Enter`**, nós **rodamos a célula** e passamos para a próxima automaticamente.
- Também podemos fazer isso pela interface, mas é mais prático usar este atalho.

## Nosso primeiro código
Rode os blocos de código abaixo com **`Shift + Enter`**:

```python
mensagem = 'Olá mundo!'
print(mensagem)
```

### Como o código funciona?

#### Texto em Python
- Para escrever textos em Python, usamos **aspas**.
- Podemos usar tanto aspas simples (`' '`) ou duplas (`" "`).
- Em Python os textos são chamados de **strings**.

#### Variáveis
- Podemos guardar **valores** dentro de **variáveis**.
- Variáveis podem guardar qualquer tipo de valor: textos, números, tabelas, figuras...
- No Google Colab, podemos ver as variáveis que criamos no botão *`{x}`* à esquerda.

#### Função `print()`
- Para exibir um texto na saída, usamos a função `print()`.
- Os parênteses indicam que **chamamos (rodamos) a função**.
- Dentro dos parênteses, passamos os **argumentos** que queremos exibir.

## Outros exemplos de código
Usando variáveis numéricas (Python usa `.` para decimal):



```python

x = 1
y = 2.5
soma = x + y

print(soma)
```

Aplicando formatação de texto:



```python
nome = 'Juliano'
idade = 32

texto = f'Meu nome é {nome} e tenho {idade} anos.'
print(texto)
```

### Formatando texto
- Usamos as chamadas **f-strings** para inserir variáveis no texto.
- Como usar:
  - Insira `f` à esquerda do string.
  - Insira variáveis dentro de chaves (`{ }`).

## Saídas automáticas
- Nos Notebooks, a última linha de uma célula é exibida automaticamente.
- Isso é útil porque:
  - Podemos omitir o `print()` na linha final.
  - Se a linha for uma tabela/gráfico, o Notebook exibe em um formato útil (veremos mais para frente).

```python
nome = 'Juliano'
idade = 32

f'Meu nome é {nome} e tenho {idade} anos.'
```

**Importante**: isso só funciona em Notebooks como os do Google Colab. Em um script Python, use `print()`.

## Para que usar Notebooks?
- Misturar trechos de texto com blocos de código.
- Testar ideias rapidamente (prototipar).
- Explorar dados e gerar gráficos.
