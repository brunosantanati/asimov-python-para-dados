## Pergunta: **qual a proporção de viagens por cada cargo**?
Para responder, usaremos o método `df.value_counts()`:

```python
df_viagens['Cargo'].value_counts()
```

Note que:
- `df.value_counts()` conta quantas ocorrências de cada valor há na coluna.
- Cada valor vai para o índice e aparece uma única vez (dimensões mudam!)
- Podemos usar `normalize=True` para obter valores em porcentagem:

```python
df_viagens['Cargo'].value_counts(normalize=True) * 100
```

Podemos dar um nome melhor para a coluna resultante com o método `df.rename()`:

```python
(df_viagens['Cargo'].value_counts(normalize=True) * 100).rename('Proporção Viagens')
```

E por fim, usar o método `df.reset_index()` para passar o índice (cargos) de volta para uma coluna padrão:

```python
(df_viagens['Cargo'].value_counts(normalize=True) * 100).rename('Proporção Viagens').reset_index()
```
