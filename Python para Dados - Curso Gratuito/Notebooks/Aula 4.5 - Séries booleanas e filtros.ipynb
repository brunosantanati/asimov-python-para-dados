```python
caminho_dados = '/content/drive/MyDrive/Curso AD/dados_viagens/2023_Viagem.csv'
```

```python
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.2f}'.format)

# Lendo os dados
df_viagens = pd.read_csv(caminho_dados, encoding='Windows-1252', sep=';', decimal=',')

# Criando nova coluna de despesas
df_viagens['Despesas'] = df_viagens['Valor diárias'] + df_viagens['Valor passagens'] + df_viagens['Valor outros gastos']
```

```python
df_viagens
```

```python
viagens_por_cargo = (df_viagens['Cargo'].value_counts(normalize=True) * 100).rename('Proporção Viagens').reset_index()
viagens_por_cargo
```

Vamos aprender agora a **filtrar** os dados agrupados.

## Séries booleanas

Fazemos **comparações em Python** usando os seguintes operadores:
- Maior que `>`
- Menor que `<`
- Maior ou igual a `>=`
- Menor ou igual a `<=`
- Igual a `==`
- Diferente de `!=`

Comparações sempre retornam um valor booleano: **verdadeiro** (`True`) ou **falso** (`False`).

```python
10 > 5
```

```python
4 == 4
```

```python
10 < 3
```

Comparações com séries do Pandas retornam uma **série booleana**, com valores verdadeiros (`True`) e falsos (`False`):

```python
viagens_por_cargo['Proporção Viagens'] > 1
```

Usamos séries booleanas para selecionar linhas/colunas. Apenas as linhas com valor `True` são retornadas.

Em outras palavras, **séries booleanas funcionam como filtros**!

```python
filtro_mais_de_1_pct = viagens_por_cargo['Proporção Viagens'] > 1
```

```python
viagens_por_cargo[filtro_mais_de_1_pct]
```

### Exemplo de filtro
Vamos filtrar as linhas cujo cargo começa com a palavra `"TECNICO"`:

```python
viagens_por_cargo['Cargo']
```

```python
viagens_por_cargo['Cargo'].str.startswith('TECNICO')
```

```python
filtro_eh_tecnico = viagens_por_cargo['Cargo'].str.startswith('TECNICO')
```

```python
viagens_por_cargo[filtro_eh_tecnico]
```

### Combinando filtros
Podemos usar os operadores abaixo para combinar séries booleanas:
- `&` retorna `True` se ambos os valores forem `True` (operador AND)
- `|` retorna `True` se pelo menos um dos valores for `True` (operador OR)
- `~` inverte o valor `True`/`False` (operador NOT)

Vamos ver o resultado em um DataFrame novo:

```python
viagens_por_cargo[filtro_eh_tecnico & filtro_mais_de_1_pct]
```

```python
viagens_por_cargo[filtro_eh_tecnico | filtro_mais_de_1_pct]
```

```python
viagens_por_cargo[~filtro_mais_de_1_pct]
```

Dá para combinar de diversas formas, mas cuidado para não gerar uma lógica muito complexa!

```python
viagens_por_cargo[filtro_eh_tecnico & (~filtro_mais_de_1_pct)]
```

## Teste!

- Quais cargos gastaram mais de R\$ 10 milhões em viagens?

```python
df_viagens.groupby('Cargo')['Despesas'].sum().reset_index()
```

```python
gastos_totais_por_cargo = df_viagens.groupby('Cargo')['Despesas'].sum().reset_index()
```

```python
gastos_totais_por_cargo[gastos_totais_por_cargo['Despesas'] > 10_000_000].sort_values(by='Despesas', ascending=False)
```

- Quais cargos tiveram uma média de gastos maior que R\$ 10 mil em viagens?

```python
gasto_medio_por_cargo = df_viagens.groupby('Cargo')['Despesas'].mean().reset_index()
```

```python
gasto_medio_por_cargo[gasto_medio_por_cargo['Despesas'] > 10_000].sort_values(by='Despesas', ascending=False)
```
