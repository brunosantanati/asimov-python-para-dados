```python
caminho_dados = '/content/drive/MyDrive/Curso AD/dados_viagens/2023_Viagem.csv'
```

```python
import pandas as pd

df_viagens = pd.read_csv(caminho_dados, encoding='Windows-1252', sep=';')
```

## Exibindo todas as colunas
Quando há muitas colunas, o pandas oculta algumas da tabela. Podemos forçá-lo a exibir as colunas dessa forma:

```python
pd.set_option('display.max_columns', None)

df_viagens
```

Poderíamos fazer o mesmo com linhas, mas são muitas.
- Não precisamos "ver" os dados a todo momento!

## Acessando colunas
Acessamos uma coluna usando colchetes (`[ ]`) e passando o nome da coluna como string.
- Atenção para passar o nome correto: espaços, letras maiúsculas/minúsculas, acentos...

```python
df_viagens['Nome do órgão superior']
```

```python
df_viagens['Valor diárias']
```

## Selecionando múltiplas colunas
Podemos usar uma **lista de strings** para selecionar algumas colunas da tabela:

```python
colunas = ['Nome do órgão superior', 'Valor diárias']
df_viagens[colunas]
```

## Operações com uma coluna
Podemos fazer operações com colunas do Pandas - essa é uma de suas principais forças! Muitas dessas operações existem na forma de **métodos**.
- Acessamos métodos usando ponto (` . `) ao final da coluna, seguido do nome do método.
- Assim como outras funções, precisamos rodar métodos com parênteses.
- Alguns métodos aceitam argumentos.

```python
df_viagens['Nome do órgão superior'].str.upper()
```

```python
df_viagens['Nome do órgão superior'].str.len()
```

Podemos encadear métodos um atrás do outro!

```python
df_viagens['Nome do órgão superior'].str.upper().str.replace('MINISTÉRIO', 'MIN.')
```

### Não decore métodos!
Não se preocupe em decorar nomes ou argumentos de métodos - isso tudo você pega com a prática!
- Pesquiso diversas vezes a forma de fazer a operação X em Python (às vezes, o link já está visitado...)
- Mais importante do que decorar é conseguir a informação rapidamente!

## Em resumo
- Use colchetes `[ ]` para selecionar colunas pelos nomes.
- Use ponto `.` para acessar operações (métodos), e parênteses `( )` para rodá-los.
- Objetos diferentes tem métodos diferentes. Alguns métodos são só de tabelas, outros só de colunas. Mas não decore!
