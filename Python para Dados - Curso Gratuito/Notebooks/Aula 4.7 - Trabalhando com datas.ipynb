## Script até aqui

```python
caminho_dados = '/content/drive/MyDrive/Curso AD/dados_viagens/2023_Viagem.csv'
```

```python
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.2f}'.format)

# Lendo os dados
df_viagens = pd.read_csv(caminho_dados, encoding='Windows-1252', sep=';', decimal=',')

# Criando nova coluna de despesas
df_viagens['Despesas'] = df_viagens['Valor diárias'] + df_viagens['Valor passagens'] + df_viagens['Valor outros gastos']

# Ajustando valores nulos na coluna de cargo
df_viagens['Cargo'] = df_viagens['Cargo'].fillna('NÃO IDENTIFICADO')
```

```python
df_viagens
```

## Datas
Temos duas colunas que representam datas:

```python
df_viagens[['Período - Data de início', 'Período - Data de fim']]
```

Mas o tipo de dado aqui é texto (string):

```python
df_viagens['Período - Data de início']
```

Posso usar `pd.to_datetime` para converter em uma data de fato dentro de Python!
- Com isso, consigo fazer operações específicas (como veremos a seguir).
- Como as datas estão no formato brasileiro, precisamos especificar o formato.
  - `%d` é o dia
  - `%m` é o mês
  - `%Y` é o ano

```python
pd.to_datetime(df_viagens['Período - Data de início'], format='%d/%m/%Y')
```

E substituindo nos dados:

```python
df_viagens['Período - Data de início'] = pd.to_datetime(df_viagens['Período - Data de início'], format='%d/%m/%Y')
df_viagens['Período - Data de fim'] = pd.to_datetime(df_viagens['Período - Data de fim'], format='%d/%m/%Y')
```

```python
df_viagens
```

## Novas colunas de datas

Com os dados no formato `datetime`, conseguimos fazer operações com datas!

```python
df_viagens['Mês da viagem'] = df_viagens['Período - Data de início'].dt.month_name()
df_viagens['Dias de viagem'] = df_viagens['Período - Data de fim'] - df_viagens['Período - Data de início']
```

```python
df_viagens
```

Ao fazermos a operação, o tipo de dado virou um "timedelta" (intervalo de tempo):

```python
df_viagens['Dias de viagem']
```

Podemos transformá-lo de volta para números de fato:

```python
df_viagens['Dias de viagem'] = df_viagens['Dias de viagem'].dt.days
df_viagens['Dias de viagem']
```
